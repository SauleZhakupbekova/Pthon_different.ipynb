```python
plt.title('График очень важных показателей')
plt.xlabel('Дни')
plt.ylabel('Стоимость')
plt.grid() # сетка на графике
# plt.subplot(2, 2, 2) # subplot(<кол-во граф.>, <граф. в строке>, <номер граф.>)
plt.plot([1, 2, 3, 4, 5], [10, 11, 14, 15, 22], 'r--') # построение графика по х и у
# plt.subplot(2, 2, 1)
plt.plot([1, 2, 3, 4, 5], [12, 13, 14, 16, 18], 'g-o')
plt.savefig('qwerty.jpg')
plt.show()
```
