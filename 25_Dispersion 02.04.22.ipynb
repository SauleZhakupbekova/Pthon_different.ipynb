```python
def pred_func(ml_input_train, ml_output_train, input_test):
    
    salary = ml_input_train['salary']
    age = ml_input_train['age']
    budget = ml_output_train['budget']
    
    result = salary / age
    score_min = 1 - min(result / budget)
    score_max = max(result / budget) - 1
    
    print('Result of dispersia: ', (score_min + score_max) / 2 * 100, '%')
    return input_test['salary'] / input_test['age']
```
