```python
dct = {
    'name': [''.join(rd.choices([*'йцукенгшщзхъёфывапролджэячсмитьбю'], k=5)) for i in range(100)],
    'salary': [rd.randint(20, 120)*1000 for i in range(100)],
    'age': [rd.randint(18, 40) for i in range(100)]
}
# dct['budget'] = [dct['salary'][i] / dct['age'][i] for i in range(len(dct['salary']))]
```

```python
def x(integer):
    return round(integer, 2)
```

```python
# df['budget_normal'] = [x(i) for i in df.budget]
df['budget'] = df['budget'].apply(x)
```

```python
def pred_func(ml_input_train, ml_output_train, input_test):
    
    salary = ml_input_train['salary']
    age = ml_input_train['age']
    budget = ml_output_train['budget']
    
    result = salary / age
    score_min = 1 - min(result / budget)
    score_max = max(result / budget) - 1
    print('Точность непопадания:  ', (score_min + score_max)/2*100, '%')
    return input_test['salary'] / input_test['age']
```

```python
input_train = df[['salary', 'age']].loc[:79]
```

```python
input_train
```

```python
output_train = df[['budget']].loc[:79]
```

```python
output_train
```

```python
input_test = df[['salary', 'age']].loc[80:]
```

```python
input_test
```

```python
pred_func(input_train, output_train, input_test)
```
